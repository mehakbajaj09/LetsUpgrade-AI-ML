 QUES 6. HOW TO KNOE THE DATES OF YESTERDAY, TODAY AND TOMORROW.

```python
import numpy as np

yesterday = np.datetime64('today') - np.timedelta64(1)
today     = np.datetime64('today')
tomorrow  = np.datetime64('today') + np.timedelta64(1)
print(yesterday)
print(today)
print(tomorrow)
```
